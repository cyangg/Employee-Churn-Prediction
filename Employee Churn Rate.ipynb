```python
import pandas as pd
import numpy as np
```

```python
path=r'D:\Kaggle\MFG10YearTerminationData.csv'
df=pd.read_csv(path)
df = df.drop(['EmployeeID'],axis=1)
df.head()
```

```python
df['gender_short']=df['gender_short'].map({'M':1,'F':0})
df['STATUS']=df['STATUS'].map({'ACTIVE':1,'TERMINATED':0})
df['job_title']=df['job_title'].replace(['Meat Cutter','Dairy Person','Produce Clerk','Baker','Cashier','Shelf Stocker',
                                        'Accounting Clerk','Benefits Admin','Accounts Receiveable Clerk','Trainer'
                                        ,'Accounts Payable Clerk'],'Low')
df['job_title']=df['job_title'].replace(['Recruiter','HRIS Analyst','Labor Relations Analyst','Auditor','Investment Analyst',
                                         'Compensation Analyst','Systems Analyst','Corporate Lawyer'],'Mid')
df['job_title']=df['job_title'].replace(['Customer Service Manager','Processed Foods Manager','Meats Manager','Bakery Manager',
                                         'Produce Manager','Store Manager','Exec Assistant, VP Stores','Exec Assistant, Legal Counsel',
                                         'CHief Information Officer','Exec Assistant, Human Resources','CEO','VP Stores',
                                         'VP Finance','VP Human Resources','Exec Assistant, Finance','Director, Recruitment',
                                         'Legal Counsel','Dairy Manager','Director, Accounting','Director, HR Technology',
                                         'Director, Accounts Payable','Director, Employee Records','Director, Accounts Receivable',
                                         'Director, Compensation','Director, Audit','Director, Labor Relations','Director, Investments',
                                         'Director, Training'],'High')
df['job_title']=df['job_title'].map({'Low':0,'Mid':1,'High':2})
df.head()
```

```python
# Train & Test data
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X=df.drop(['STATUS','recorddate_key','birthdate_key','orighiredate_key','terminationdate_key',
          'city_name','department_name','gender_full','termreason_desc',
          'termtype_desc','BUSINESS_UNIT'],axis=1)
y=df['STATUS']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
for i in [X_train, X_test, y_train, y_test]:
    i.index = range(i.shape[0])



smote = SMOTE(random_state = 101)
X_oversample, y_oversample = smote.fit_resample(X_train, y_train)
```

```python
# DT
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_oversample, y_oversample)
y_pred = clf.predict(X_test)
acc_DT=accuracy_score(y_test,y_pred)
print("Accuracy: {:.2%}".format(acc_DT))
```
